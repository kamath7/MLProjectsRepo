```python
from scipy.stats import norm
import numpy as np 

x = np.arange(-3,3,0.0003)

plt.plot(x, norm.pdf(x))
```

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0,0.5))
```

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0,0.5))
plt.savefig("mygraph.png",format="png")
```

```python
axes = plt.axes()
axes.set_xlim([-5,5])
axes.set_ylim([0,1.0])
axes.set_xticks([0.2,0.1,0])
axes.set_yticks([0,0.1,0.2,0.3])
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x,1.0,0.5))
```

```python
axes = plt.axes()
axes.set_xlim([-5,5])
axes.set_ylim([0,1.0])
axes.set_xticks([0.2,0.1,0])
axes.set_yticks([0,0.1,0.2,0.3])
axes.grid()
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x,1.0,0.5))
```

```python
axes = plt.axes()
axes.set_xlim([-5,5])
axes.set_ylim([0,1.0])
axes.set_xticks([0.2,0.1,0])
axes.set_yticks([0,0.1,0.2,0.3])
axes.grid()

plt.plot(x, norm.pdf(x), 'b-')
plt.plot(x, norm.pdf(x,1.0,0.5), 'x:')
plt.legend(['Nice','Shoes'], loc=4)
```

```python
values = [12,55,22,3,12]
colors = ['y','g','b','c','m']
explode = [0,0,0.2,0,0]
labels = ['Ferrari','Fiat','Hindustan Motors','Tata','Mahindra']
plt.pie(values, colors=colors, labels=labels,explode=explode)
plt.title("Cars")
plt.show()
```
