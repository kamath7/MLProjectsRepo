```python
import pandas as pd
```

```python
t_cols = ["user_id","movie_id","rating"]
ratings = pd.read_csv("./ml-100k/u.data",sep = "\t",names = t_cols, usecols=range(3))
m_cols = ["movie_id","title"]
movies = pd.read_csv("./ml-100k/u.item",sep="|",names=m_cols, usecols=range(2),encoding='latin-1')
```

```python
ratings = pd.merge(movies, ratings)
```

```python
ratings
```

```python
userRatings = ratings.pivot_table(index=["user_id"], columns=['title'], values='rating')
```

```python
userRatings
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method="pearson", min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.loc[7].dropna()
```

```python
myRatings
```

```python
similarFellas = pd.Series()
for i in range(0, len(myRatings.index)):
    sims = corrMatrix[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings[1])
    similarFellas = similarFellas.append(sims)

similarFellas.sort_values(inplace=True, ascending=False)
print(similarFellas.head(10))
```

```python
similarFellas = similarFellas.groupby(similarFellas.index).sum()
```

```python
similarFellas.sort_values(inplace=True, ascending=False)
print(similarFellas.head(10))
```

```python
filterSims = similarFellas.drop(myRatings.index)
filterSims.head()
```
